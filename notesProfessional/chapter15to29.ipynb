```python
while True:
    for i in range(1,5):
        print(i)
        if i == 2:
            break 
```

```python
def break_loop():
    for i in range(1, 5):
        if (i == 2):
            return(i)
        print(i)
    return(10)

break_loop()
```

```python
def break_all():
    for j in range(1, 5):
        for i in range(1,4):
            if i*j == 6:
                return(i)
        print(i*j)
break_all()
```
